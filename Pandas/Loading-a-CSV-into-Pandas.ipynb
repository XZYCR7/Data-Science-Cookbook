## Loading a CSV into Pandas

### Import Modules

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Load a CSV into Pandas

We are going to use the `read_csv` function to read from the Titanic training dataset. This function sotres the content of the `titanic-train.csv` dataset on to a variable call `df`. We call it `df` because the type of this variables is `DataFrame` as you can see. As you can see is a Pandas core frame `DataFrame`.

```python
df = pd.read_csv('../data/titanic-train.csv')
```

Since it is a dataframe it is exposed to some usefull methods. For example the `head` method. This method shows the first five lines of a `DataFrame`. You can think of a `DataFrame` as a Excel spreadsheet for Python. It is an object with columns with column names and rows indexed with an index.

```python
df.head()
```

By asiking for `info` on the `DataFrame`, we get a summary of all the features contained in the data frame, with the name of the columns, with how many non-null values we have in each column. You can see that we don't know the age of a few passangers and we don't know the cabin of many passangers. We also get usefull information about the type of data contained in that column. When we see the `object` type means that the variable is a string.

```python
df.info()
```

If we use the `describe` method we get some summary statistics about the numerical columns.

```python
df.describe()
```

### Creating a dataframe (that will be importing)

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', 'Jacobson', ".", 'Milner', 'Cooze'],
        'age': [42, 52, 36, 24, 73],
        'preTestScore': [4, 24, 31, ".", "."],
        'postTestScore': ["25,000", "94,000", 57, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'preTestScore', 'postTestScore'])
df
```

### Save a dataframe as csv in the working directory

```python
df.to_csv('../data/example.csv')
```

### Load a csv while specifying "." as missing values

```python
df = pd.read_csv('../data/example.csv', na_values=['.'])
pd.isnull(df)
```

### Load a csv while specifying "." and "NA" as missing values in the Last Name column and "." as missing values in Pre-Test Score column

```python
sentinels = {'Last Name': ['.', 'NA'], 'Pre-Test Score': ['.']}
```

```python
df = pd.read_csv('../data/example.csv', na_values=sentinels)
df
```

### Load a csv while skipping the top 3 rows

```python
df = pd.read_csv('../data/example.csv', na_values=sentinels, skiprows=3)
df
```

### Load a csv while interpreting "," in strings around numbers as thousands seperators

```python
df = pd.read_csv('../data/example.csv', thousands=',')
df
```
