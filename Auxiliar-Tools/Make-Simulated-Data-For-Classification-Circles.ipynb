## Create Simulated Data

Sicit learn is the Python's primary machine learning library and it comes with a lot of utilities like for example a dataset generator.

```python
from sklearn.datasets.samples_generator import make_circles
```

We import `make_circles`. Now we should call the `make_circles` function with some parameters. 

- `n_samples` the number of samples
- `noise` certain level of noise
- `factor` we will see what it means
- `random_state` we will see what it means

```python
X, y = make_circles(n_samples=1000, 
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

This function has two outputs, we assign them to the variables `X` and `y`. 

```python
X
```

`X` is a matrix with rows and columns. And if we look at the shape:

```python
X.shape
```

It has a thousand rows and two columns. This correspond to the number of samples `n_samples` and we have two features for each sample.

Let's plot the values of `X` and let's plot them in two different colors. In blue for the points of `X` that correspond to `y` equals to zero `X[y==0, 1]`. And in red with the cross for the points that correspond to the labels equals to one. 

```python
plt.figure(figsize=(5,5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

This is our trainng data. We have data in the outer circle, that belongs to a certain class and data in the inner space that belgons to the other class. 

In machine learning we often want to train models to separate tow classes. In this case a model should learn a kind of circular bander between the red crosses and the blue dots.
